# Import Required Libraries

```python
import pandas as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

```

#

# Upload the CSV file

```python
df = pd.read_csv('/content/canada_per_capita_income.csv')
```

```python
df.head()
```

```python
print(df.shape)
```

```python
print(df.info())
```

# Visualize the Data

```python
plt.scatter(df['year'], df['per capita income (US$)'])
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.title('Canada Per Capita Income')
plt.show()
```

# Prepare Features and Target

```python
X = df[['year']]
y = df['per capita income (US$)']
```

# Create and Train the Model

```python
model = LinearRegression()

model.fit(X, y)
```

# Predict Income for 2020

```python
prediction = model.predict([[2020]])

print("Predicted Per Capita Income in 2020:")
print(prediction[0])
```

# View Model Parameters

```python
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```

# Plot Regression Line

```python
plt.scatter(df['year'], df['per capita income (US$)'])

plt.plot(
    df['year'],
    model.predict(df[['year']]),
    color='red'
)

plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.title('Linear Regression Fit')
plt.show()
```

# Predict Multiple Future Years

```python
future_years = pd.DataFrame({
    'year': [2020, 2021, 2022, 2023, 2024]
})

predictions = model.predict(future_years)

result = pd.DataFrame({
    'Year': future_years['year'],
    'Predicted Income': predictions
})

print(result)
```
